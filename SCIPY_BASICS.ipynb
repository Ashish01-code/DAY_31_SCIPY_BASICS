```python
import numpy as np
from scipy.sparse import csr_matrix
dense_matrix = np.array([
    [0, 0, 3, 0],
    [0, 0, 0, 0],
    [5, 0, 0, 0],
    [0, 0, 2, 0]
])

print(dense_matrix)
sparse_matrix=csr_matrix(dense_matrix)
print("Data:", sparse_matrix.data)
print("INDICES:",sparse_matrix.indices)
print("Indptr:",sparse_matrix.indptr)
```

```python
import numpy as np
from scipy.optimize import minimize
def loss_function(x):
    return (x-3)**2
result=minimize(loss_function,x0=0)  
print(result)
print("MINIMUM AT X =",result.x)
print("MINIMUM VALUE =",result.fun)
```

```python
from scipy.optimize import root
def equation(x):
    return x**2-4
solution=root(equation,x0=1)
print(solution)
print("Root : ",solution.x)
print("Function value",equation(solution.x))
```
